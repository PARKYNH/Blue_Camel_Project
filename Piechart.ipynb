```python
import matplotlib.pyplot as plt

#ratio = [34, 32, 16, 18]
ratio = [100, 200, 300, 500]
labels = ['Seoul', 'Daejeon', 'Busan', 'Incheon']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
import matplotlib
matplotlib.matplotlib_fname()
```

```python
local_count_sorted = local_count.sort_values(by=[""], ascending=False)
local_count_top5 = local_count_sorted[:5]
local_count_top5
```

```python
ratio = local_count_top5.
labels = local_count_top5.
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0',
          '#ff9999', '#ffc000', '#8fd9b6', '#d395d0',
          '#ff9999', '#ffc000', '#8fd9b6', '#d395d0',
          '#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#ff9999']

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, \
        counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```
